## 4°) La conjecture Bachet de Méziriac 

```python
for a in range(45):
    for b in range(45):
        for c in range (45):
            for d in range (45):
                if (a**2+b**2+c**2+d**2 == 2020):
                    print(a, b, c, d)
```

```python
# seconde version du programme, dès qu'il trouve une solution
a, trouve = 0, False
while (a <= 45 and not trouve):
    a=a+1
    b=0
    while (b <= 45 and not trouve):
        b=b+1
        c=0
        while (c <= 45 and not trouve):
            c=c+1
            d=0
            while (d <= 45 and not trouve):
                d=d+1
                if (a**2+b**2+c**2+d**2 == 2020):
                    print("trouvé", a, b, c, d)
                    trouve = True
```

```python
for i in range(4,101):
    trouve = False
    for a in range(11):
        for b in range(11):
            for c in range(11):
                for d in range(11):
                    if (a**2+b**2+c**2+d**2 == i) and (not trouve):
                        print(f'trouvé : {a}^2 + {b}^2 + {c}^2 + {d}^2 = {i}')
                        trouve = True
    if (not trouve): print(f'AUCUNE VALEUR TROUVEE POUR {i}')
```

```python
for i in range(4,101):
    i=i+1
    a, trouve = 0, False
    while (a**2 < i and not trouve):
        b=0
        while (a**2+b**2 <= i and not trouve):
            c=0
            while (a**2+b**2+c**2 <= i and not trouve):
                d=0
                while (a**2+b**2+c**2+d**2 <= i and not trouve):
                    if (a**2+b**2+c**2+d**2 == i):
                        print(f'trouvé : {a}^2 + {b}^2 + {c}^2 + {d}^2 = {i}')
                        trouve = True
                    d = d+1
                c, d = c+1, 0

            b, c = b+1, 0
        a, b = a+1, 0

    if (not trouve): print(f'AUCUNE VALEUR TROUVEE POUR {i}')
```
